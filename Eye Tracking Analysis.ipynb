```python
Excel_High={}
Excel_Low={}

for i in excel_data.keys():
    if i in Names['High Name']:
        Excel_High[i] = excel_data[i]
    elif i in Names['Low Name']:
        Excel_Low[i] = excel_data[i]
```

```python
def average(Data, Block_No):
    No = Block_No
    n = int(len(Data)/No)
    df = [Data[i*n:i*n+n] for i in range(0, No)]
    for i in range(No):
        for j in range(len(df[i])):
            if df[i].loc[n*i + j]['Tracking Ratio'] < 70:
                df[i] = df[i].drop(n*i + j,axis=0,inplace=False)
    for j in range(No):
        df[j] = df[j].drop('Recorded',axis=1,inplace=False)
    #     df[j] = df[j].drop('Rec Pre',axis=1,inplace=False)
    ave_list = []

    column_names = df[0].columns.to_list()
    column_names.append('No')
    Average = pd.DataFrame(columns=column_names)
    
    for j in range(No):
        List = list(df[j].mean())
        List.append(int(len(df[j])))
        Average.loc[j] = List
        

    return Average
```

```python
No = 4
```

```python
Res_High = {}
for i in Excel_High.keys():
    print(i)
    Res_High[i] = average(Excel_High[i], No)
```

```python
Res_Low = {}
for i in Excel_Low.keys():
    print(i)
    Res_Low[i] = average(Excel_Low[i], No)
```

```python
Result = {**Res_High,**Res_Low}
```

```python
Dfs = [Result[i] for i in Result.keys()]
```

```python
Final = pd.concat([i for i in Dfs])
```

```python

high_len = len(Res_High.keys())
low_len = len(Res_Low.keys())
```

```python
Final['Group'] = [1 for i in range(No*high_len)] + [2 for i in range(No*low_len)]
```

```python
Final['Block'] = [i+1 for i in range(No)] * (high_len + low_len)
```

```python
Names = []
for i in Result.keys():
    for j in range(No):
        Names.append(i)
```

```python
Final['Name'] = Names
```

```python
Final_Results = {}

for i in range(len(Final.columns)):
    print(f'\033{Final.columns[i]}\033[0m')
    if Final.columns[i] == 'Group':
        break
    aov = pg.mixed_anova(dv=Final.columns[i], between='Group',
                      within='Block', subject='Name', data=Final)
    print('\n')
    print(aov)
    print('\n')
    print('\n')
    Final_Results[Final.columns[i]] = aov
```

```python
Dfss = [Final_Results[i] for i in Final_Results.keys()]
Finally = pd.concat([i for i in Dfss])
```

```python
Fac = []
for i in Final_Results.keys():
    for j in range(3):
        Fac.append(i)
```

```python
Finally['Factors'] = Fac
```
