```python
def Average_Pupil(Ind):
    
    Ind_dict = {}
    GP = Ind.groupby('Trial')
    for category,group in GP:
        Ind_dict[int(category[-3:])] = group

    #Emotion List of trials 
    
            
    Emotion_LOT_1=[]
    for i in range(4):
        for j in range(60):
            if i == 0 :
                Emotion_LOT_1.append(4+4*j)
            elif i == 1:
                Emotion_LOT_1.append(246 + 4*j)
            elif i == 2:
                Emotion_LOT_1.append(489 + 4*j)
            elif i == 3:
                Emotion_LOT_1.append(731 + 4*j)

    not_recorded = ['no' for i in Emotion_LOT_1]
    Emotion_LOT = []
    for ii in range(len(Emotion_LOT_1)):
        if Emotion_LOT_1[ii] in Ind_dict.keys():
            not_recorded[ii] = 'yes'
            Emotion_LOT.append(Emotion_LOT_1[ii])
    # Fixation count lsit for emotions for eyes
    
    average_pupil= [0 for i in range(len(Emotion_LOT_1))]
    average_pupil_minus= [0 for i in range(len(Emotion_LOT_1))]
    # Fixation count for eye
    for i in range(len(Emotion_LOT_1)):
        if Emotion_LOT_1[i] in Emotion_LOT:
    #         print(Emotion_LOT[i])
            trial = Ind_dict[Emotion_LOT_1[i]]
            trial['Average Pupil Diameter [mm]'] =  trial['Average Pupil Diameter [mm]'].replace('-','0')
            trial['Fixation Duration [ms]'] =  trial['Fixation Duration [ms]'].replace('-','0')
            trial['Average Pupil Diameter [mm]'] = [float(j) for j in trial['Average Pupil Diameter [mm]']]
            SUM_List = []
            for jj in range(len(trial['Average Pupil Diameter [mm]'])):
    #             print(jj)
    #             print(trial['Average Pupil Diameter [mm]'][jj])
                No = trial['Average Pupil Diameter [mm]'].iloc[jj] * trial['Fixation Duration [ms]'].iloc[jj] 
                SUM_List.append(No)
            SUM = np.sum(SUM_List)
            Count = np.sum(trial['Fixation Duration [ms]'])
            average_pupil[i] = SUM/Count
    
    for ii in range(len(Emotion_LOT_1)):
        if Emotion_LOT_1[ii] in Emotion_LOT:
            if Emotion_LOT_1[ii]-1 in Ind_dict.keys():
                trial = Ind_dict[Emotion_LOT_1[ii]-1]
                trial['Average Pupil Diameter [mm]'] =  trial['Average Pupil Diameter [mm]'].replace('-','0')
                trial['Fixation Duration [ms]'] =  trial['Fixation Duration [ms]'].replace('-','0')
                trial['Average Pupil Diameter [mm]'] = [float(j) for j in trial['Average Pupil Diameter [mm]']]
                SUM_List = []
                for jj in range(len(trial['Average Pupil Diameter [mm]'])):
                    No = trial['Average Pupil Diameter [mm]'].iloc[jj] * trial['Fixation Duration [ms]'].iloc[jj] 
                    SUM_List.append(No)
                SUM = np.sum(SUM_List)
                Count = np.sum(trial['Fixation Duration [ms]'])
                average_pupil_minus[ii] = average_pupil[ii] - (SUM/Count)
            
        


    return average_pupil,average_pupil_minus, not_recorded
```

```python
def Fix_Face(Ind, Emotion_List,Face_Co):
    
    Ind_dict = {}
    GP = Ind.groupby('Trial')
    for category,group in GP:
        Ind_dict[int(category[-3:])] = group
     #Dropping NAN from Emotion_List
    Emotion_List = list(Emotion_List.dropna())
    #Emotion List of trials 
    
            
    Emotion_LOT_1=[]
    for i in range(4):
        for j in range(60):
            if i == 0 :
                Emotion_LOT_1.append(6+4*j)
            elif i == 1:
                Emotion_LOT_1.append(248 + 4*j)
            elif i == 2:
                Emotion_LOT_1.append(491 + 4*j)
            elif i == 3:
                Emotion_LOT_1.append(733 + 4*j)

    Emotion_LOT = []
    Trial_No = []
    for ii in range(len(Emotion_LOT_1)):
        Trial_No.append(ii)
        if Emotion_LOT_1[ii] in Ind_dict.keys():
            Emotion_LOT.append(Emotion_LOT_1[ii])
            
    Fix_face  = [0 for i in range(len(Emotion_LOT_1))]
    Time_face = [0 for i in range(len(Emotion_LOT_1))]
    for i in range(len(Emotion_LOT_1)):
        if Emotion_LOT_1[i] in Emotion_LOT:
        
            Fix_counter_face = 0
            Fix_timer_face = 0
            emotion = Emotion_List[Trial_No[i]]
            
            trial = Ind_dict[Emotion_LOT_1[i]]
            
            Min_X = Face_Co[emotion]['Xpos'][0]
            Dia_X = Face_Co[emotion]['Xpos'][1]/2
            Min_Y = Face_Co[emotion]['Ypos'][0]
            Dia_Y = Face_Co[emotion]['Ypos'][1]/2
            
            for j in range(len(trial)):
                
                x = trial.iloc[j]['Position X']
                y = trial.iloc[j]['Position Y']
                Oval_dia_x = ((x-Min_X-Dia_X)/Dia_X)**2
                Oval_dia_y = ((y-Min_Y-Dia_Y)/Dia_Y)**2
                Eq = Oval_dia_x + Oval_dia_y
                if  Eq <= 1 :
                    Fix_counter_face += 1
                    Fix_timer_face += trial.iloc[j]['Fixation Duration [ms]']
                    
            Fix_face[i] = Fix_counter_face
            Time_face[i] = Fix_timer_face
        

    return Fix_face,Time_face
```

```python

def Firsts(Ind, Emotion_List,AOI_eye, AOI_mouth):
    
    Ind_dict = {}
    GP = Ind.groupby('Trial')
    for category,group in GP:
        Ind_dict[int(category[-3:])] = group

    #Dropping NAN from Emotion_List
    Emotion_List = list(Emotion_List.dropna())
    #Emotion List of trials 
    
            
    Emotion_LOT_1=[]
    for i in range(4):
        for j in range(60):
            if i == 0 :
                Emotion_LOT_1.append(6+4*j)
            elif i == 1:
                Emotion_LOT_1.append(248 + 4*j)
            elif i == 2:
                Emotion_LOT_1.append(491 + 4*j)
            elif i == 3:
                Emotion_LOT_1.append(733 + 4*j)
    Emotion_LOT = []
    Trial_No = []
    for ii in range(len(Emotion_LOT_1)):
        Trial_No.append(ii)
        if Emotion_LOT_1[ii] in Ind_dict.keys():
            Emotion_LOT.append(Emotion_LOT_1[ii])
            
    # Fixation count lsit for emotions for eyes
    
    # Fixation count for eye
    First_Fix_Eyes = [0 for i in range(240)]
    First_Fix_Eyes_Time = [0 for i in range(240)]
    for i in range(len(Emotion_LOT_1)):
        if Emotion_LOT_1[i] in Emotion_LOT:
            First_Eye = 0
            First_Eye_Time = 0
            emotion = Emotion_List[Trial_No[i]]
    
            # AOI_eye
            Min_X = AOI_eye[emotion]['Xpos'][0]
            Max_X = AOI_eye[emotion]['Xpos'][1]
            Min_Y = AOI_eye[emotion]['Ypos'][0]
            Max_Y = AOI_eye[emotion]['Ypos'][1]
            
            #  determining the trial we are counting
            
            trial = Ind_dict[Emotion_LOT_1[i]]
            
            for j in range(len(trial)):
                if Min_X < trial.iloc[j]['Position X'] < Max_X and Min_Y < trial.iloc[j]['Position Y'] < Max_Y:
                    First_Eye += j+1
                    First_Eye_Time = trial.iloc[j]['Fixation Duration [ms]']
                    break
                    
            First_Fix_Eyes[i] = First_Eye
            First_Fix_Eyes_Time[i] = First_Eye_Time
        
            # Fixation count lsit for emotions for eyes
    First_Fix_Mouths = [0 for i in range(240)]
    First_Fix_Mouths_Time = [0 for i in range(240)]
    # Fixation count for eye
    for i in range(len(Emotion_LOT_1)):
        if Emotion_LOT_1[i] in Emotion_LOT:
            First_Mouth = 0
            First_Mouth_Time = 0
            emotion = Emotion_List[Trial_No[i]]
    
            # AOI_eye
            Min_X = AOI_mouth[emotion]['Xpos'][0]
            Max_X = AOI_mouth[emotion]['Xpos'][1]
            Min_Y = AOI_mouth[emotion]['Ypos'][0]
            Max_Y = AOI_mouth[emotion]['Ypos'][1]
            
            #  determining the trial we are counting
            
            trial = Ind_dict[Emotion_LOT_1[i]]
            for j in range(len(trial)):
                if Min_X < trial.iloc[j]['Position X'] < Max_X and Min_Y < trial.iloc[j]['Position Y'] < Max_Y:
                    First_Mouth = j + 1
                    First_Mouth_Time = trial.iloc[j]['Fixation Duration [ms]']
                    break
            First_Fix_Mouths[i] = First_Mouth
            First_Fix_Mouths_Time[i] = First_Mouth_Time
        
    return First_Fix_Eyes,First_Fix_Eyes_Time,First_Fix_Mouths,First_Fix_Mouths_Time
            
    
        
        
```

```python
def eye_tracking_measures(Ind, Emotion_List,AOI_eye, AOI_mouth,Face_Co):
    Data_1,Data_2 = Fix_Counter(Ind, Emotion_List,AOI_eye, AOI_mouth)
    Data_3,Data_4 = Fix_Time(Ind, Emotion_List,AOI_eye, AOI_mouth)
    Data_5,Data_6,Data_15 = Average_Pupil(Ind)
    Data_16,Data_17,Data_18 = Average_Pupil_Last(Ind)
    Data_7,_Data_8 = Fix_Face(Ind, Emotion_List,Face_Co)
    Data_9,Data_10,Data_11,Data_12 = Firsts(Ind, Emotion_List,AOI_eye, AOI_mouth)
    Data_13,Data_14 = Total(Ind, Emotion_List)
    Data = zip(Data_1,Data_2,Data_3,Data_4,Data_5,Data_6,Data_7,_Data_8,Data_9,Data_10,Data_11,Data_12,Data_13,Data_14,Data_15, Data_16,Data_17,Data_18)
    df = pd.DataFrame(Data,columns=['Fixation Count Eye','Fixation Count Mouth',
                                   'Fixation Time Eye','Fixation Time Mouth',
                                   'Average Pupil For Trial','Average Pupil Minus Previous Average',
                                   'Fixation Count Face','Fixation Time Face',
                                   'First Fix Eye','First Fix Eye Time','First Fix Mouth',
                                   'First Fix Mouth Time','Total Fixation Count','Total Fixation Time','Recorded',
                                    'Ave Pre','Ave Min Pre', 'Rec Pre'])
    return df
```
